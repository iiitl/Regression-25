```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/wines_SPA.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['type'].fillna(df['type'].mode()[0],inplace=True)
df['acidity'].fillna(df['acidity'].mean(),inplace=True)
df['body'].fillna(df['body'].mean(),inplace=True)
```

```python
df=df.drop_duplicates()
```

```python
df['year']=df['year'].dropna()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
df['wine']=LabelEncoder().fit_transform(df['wine'])
df['type']=LabelEncoder() .fit_transform(df['type'])
df['winery']=LabelEncoder() .fit_transform(df['winery'])
df['region']=LabelEncoder() .fit_transform(df['region'])
df['country']=LabelEncoder() .fit_transform(df['country'])
```

```python
df["year"].nunique()
```

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
```

```python
df['year'].dtypes
```

```python
df['year'].fillna(df['year'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
new_df=df.drop(['price'],axis=1,inplace=False)
target=df["price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_df, target, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

model.fit(X_train, y_train,  validation_data=(X_test, y_test))

predictions = model.predict(X_test)

loss,mae=model.evaluate(X_test, y_test)

print(f"mean absolute error : {mae}")
print(f"Loss : {loss}")
```
