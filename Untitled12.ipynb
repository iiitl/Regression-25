```python
import warnings
import matplotlib.pyplot as plt

# Ignore warnings for clean output
warnings.filterwarnings('ignore')

```

```python
# Ensure column names are clean
df.columns = df.columns.str.strip()  # Remove leading/trailing spaces
df.columns = df.columns.str.replace(r'[^A-Za-z0-9_]', '_', regex=True)  # Replace special characters
```

```python
# Use ggplot style for aesthetics
plt.style.use('ggplot')
```

```python
# Loop through numeric columns
for i in df.select_dtypes(include=['int64','float64']).columns:
    plt.figure(figsize=(8, 5))  # Set figure size
    plt.hist(df[i], bins=15, color='skyblue', edgecolor='black', alpha=0.7,rwidth=1.0)  # Add color, edges, transparency
    plt.title(f'Distribution of {i}', fontsize=14)  # Add title
    plt.xlabel(i, fontsize=12)  # Label x-axis
    plt.ylabel('year', fontsize=12)  # Label y-axis
    plt.grid(True, alpha=0.3)  # Add light grid for readability
    plt.tight_layout()  # Adjust layout
    plt.show()
```

```python
#distplot
import warnings
warnings.filterwarnings('ignore')
plt.style.use('ggplot')
for i in df.select_dtypes(include=['int64','float64']).columns:
  sns.distplot(df[i])
  plt.show()
```

```python
#Boxplot to identify outliers
import warnings
warnings.filterwarnings('ignore')
plt.style.use('ggplot')
for i in df.select_dtypes(include=['int64','float64']).columns:
  sns.boxplot(data=df,x=i)
  plt.show()
```

```python
#correlation with heatmap to interpret the realation and multicollinearity
s= df.select_dtypes(include=['int64','float64']).corr()
plt.figure(figsize=(15,15))
sns.heatmap(s, annot=True, cmap='coolwarm')
plt.show()
```
