We create a function that does our work of train test split. It takes independent variables dataframe,dependent variable dataframe,split size as input.
It give xtrain,xtest,ytrain,ytest as output


```python
import pandas as pd
def testtrainsplit(df,test,test_size):
  #now we concat both the dataframe to preserve conection while shuffling
  new_df=df;
  new_df["test"]=test;
  #now we shuffle the newdataframe
  new_df=new_df.sample(frac=1).reset_index(drop=True);

  #now we take random value according to test_size
  test_df=new_df.sample(frac=test_size).reset_index(drop=True);
  train_df=new_df.drop(test_df.index).reset_index(drop=True);

  return train_df.drop(columns=["test"]),test_df.drop(columns=["test"]),train_df["test"],test_df["test"];



```

```python
df=pd.read_csv("/content/wines_SPA.csv")
df.head()
```

```python
Xtrain,Xtest,Ytrain,Ytest=testtrainsplit(df.drop(columns=["price"]),df["price"],0.2)
```
