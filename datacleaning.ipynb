```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/wines_SPA.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['type'].fillna(df['type'].mode()[0],inplace=True)
df['acidity'].fillna(df['acidity'].mean(),inplace=True)
df['body'].fillna(df['body'].mean(),inplace=True)
```

```python
df=df.drop_duplicates()
```

```python
df['year']=df['year'].dropna()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
df['wine']=LabelEncoder().fit_transform(df['wine'])
df['type']=LabelEncoder() .fit_transform(df['type'])
df['winery']=LabelEncoder() .fit_transform(df['winery'])
df['region']=LabelEncoder() .fit_transform(df['region'])
df['country']=LabelEncoder() .fit_transform(df['country'])
```

```python
df["year"].nunique()
```

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
```
