```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/synthetic_linear_regression.csv")
df.head()
```

```python
plt.plot(df["Feature"],df['Target'])
```

```python
#initialising slope and intercept to zero
#slope =m intercept =c
m=0.0
c=0.0
learning_rate=0.00001
iter=1000
n=len(df)
```

```python
while(iter):
    y_pred = m * df["Feature"] + c
    error = y_pred - df['Target']

    m_grad=(2/n)*np.sum(df["Feature"]*error)
    c_grad=(2/n)*np.sum(error)

    m-=learning_rate * m_grad
    c-=learning_rate * c_grad

    iter=iter-1
```

```python
df["pred"]= m*df['Feature'] + c
```

```python
print(m)
print(c)
```

```python
plt.plot(df["Feature"],df['Target'])
plt.plot(df["Feature"],df['pred'])
plt.show()
```
