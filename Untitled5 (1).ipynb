```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("wines_SPA.csv")
df.shape
```

```python
dup=df.duplicated()
df[dup]
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
miss=df.isnull().sum()
miss
```

```python
df.drop('country',axis=1,inplace=True)
df.shape
```

```python
#list of unique values of body and acidity based on unique values of type column
unique_types = df['type'].unique()
result=[]
for wine_type in unique_types:
  type_df=df[df['type']==wine_type]
  unique_body=type_df['body'].unique()
  unique_acidity=type_df['acidity'].unique()
  result.append({
      'type':wine_type,
      'unique_body':unique_body,
      'unique_acidity':unique_acidity
  })
result
```

```python
# prompt: wherever body or acidity is empty fill values based with arr[0] in above list

# Fill missing 'body' values
for item in result:
    wine_type=item['type']
    unique_body_values=item['unique_body']

    if unique_body_values.size > 0:
        body_fill_value = unique_body_values[0]
        df.loc[(df['type']==wine_type) & (df['body'].isnull()),'body']=body_fill_value

for item in result:
    wine_type=item['type']
    unique_acidity_values=item['unique_acidity']

    if unique_acidity_values.size > 0:
        acidity_fill_value = unique_acidity_values[0]
        df.loc[(df['type'] == wine_type) & (df['acidity'].isnull()), 'acidity'] = acidity_fill_value
        miss=df.isnull().sum()
miss
```

```python
#values of body and acidity when type=NA -- FROM THIS WE UNDERSTAND THAT WHEN type is NA then body and acidity are also NA
na_df=df[df['type']=='NA']
unique_bodies=na_df['body'].unique()
unique_acidity=na_df['acidity'].unique()

print("body values", unique_bodies)
print("acidity values", unique_acidity)
```

```python

# Filling missing 'type',body and acidity values with mode since not many NA values
df['type'].fillna(df['type'].mode, inplace=True)
df['body'].fillna(df['body'].mode, inplace=True)
df['acidity'].fillna(df['acidity'].mode, inplace=True)

miss = df.isnull().sum()
miss
```

```python
df['year'].fillna(df['year'].mode()[0], inplace=True)

miss = df.isnull().sum()
miss
df.shape

```
